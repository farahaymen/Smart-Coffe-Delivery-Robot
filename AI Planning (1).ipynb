```python
def update_locations(location, TA):
    for loc in Map:
        if Map[loc] == TA:
            Map[loc] = 'NA'
        if loc == location:
            Map[location] = TA
    return loc
```

```python
def pickup_coffee(orders, cups_available, cups_carried):
    for order in orders:
        for cup in cups_available:
            if order[1] == cup and len(cups_carried) < 4:
                cups_carried.append(cup)
                cups_available.remove(cup)
            else:
                print('Can\'t carry anymore.')
    return cups_available, cups_carried
```

```python
from collections import deque

class Graph:
    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        return self.adjacency_list[v]
    
    def co(self, n):
        co = { 
        'CoffeeMaker': (1, 2),
        'CCoffeeDw': (5, 1),
        'Corridor1': (6, 1),
        'Corridor2': (6, 3),
        'CCRDw': (4, 3),
        'Classroom': (2, 3),
        'CTADw1': (4, 5),
        'CTADw2': (9, 3),
        'Area1': (5, 7),
        'Area2': (5, 10),
        'Area3': (8, 10),
        'Area4': (9, 8),
        'Area5': (10, 4),
        'T1': (11, 3),
        'T2': (11, 4),
        'T3': (11, 5),
        'T4': (11, 7),
        'T5': (11,8),
        'T6': (11, 9),
        'T7': (11, 10),
        'T8': (10, 10),
        'T9': (10, 9),
        'T10': (10, 8),
        'T11': (10, 7),
        'T12': (9, 11),
        'T13': (8, 12),
        'T14': (6, 12),
        'T15': (6, 11),
        'T16': (5, 11),
        'T17': (5, 12),
        'T18': (4, 12),
        'T19': (3, 11),
        'T20': (6, 9),
        'T21': (6, 8),
        'T22': (6, 7),
        'T23': (6, 6),
        'T24': (8, 6),
        'T25': (8, 7),
        'T26': (8, 8),
        'T27': (8, 9),
        'T28': (1, 3)
        }
        return co[n]
    
    def h(self, end_node, n):
        H={}
        for i in self.adjacency_list:
             H[i] = abs(self.co(i)[0]-self.co(end_node)[0]) + abs(self.co(i)[1]-self.co(end_node)[1])
            
        return(H[n])
       
    def deliver_coffee(self, graph, start_node, end_node, orders, cups_available, cups_carried, screen):
        
        cups_available1 = cups_available
        cups_carried1 = cups_carried
    
        clock = pygame.time.Clock()
        open_list = set([start_node])
        closed_list = set([])

        cups_available1, cups_carried1 = pickup_coffee(orders, cups_available, cups_carried)
            # g contains current distances from start_node to all other nodes
            # the default value (if it's not found in the map) is +infinity
        g = {}

        g[start_node] = 0
        sub = 'Dw'
            # parents contains an adjacency map of all nodes
        parents = {}
        parents[start_node] = start_node
        #graph = Graph
        
        while len(open_list) > 0:
            n = None

                # find a node with the lowest value of f() - evaluation function
            for v in open_list:
            
                    
                if n == None or g[v] + graph.h(end_node,v) < g[n] + graph.h(end_node,n):
                    n = v
                                
            if n == None:
                print('Path does not exist!')
                return None

                # if the current node is the stop_node
                # then we begin reconstructin the path from it to the start_node
            if n == end_node:
                
                for loc in Map:
                    if loc == n:
                        for order in orders:
                            if order[0] == Map[loc]:
                                cups_carried1.remove(order[1])
                                orders.remove(order)
                reconst_path = []

                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]
                    
                            
                reconst_path.append(start_node)
                print('Path found: {}'.format(reconst_path[::-1]))
                
                for n in reconst_path[::-1]:
                    pygame.draw.circle(screen, "Green", (self.co(n)[0]*80+500, self.co(n)[1] * 80), 20)
                    pygame.display.update()
                    clock.tick(2)
                    
                    if sub in n:
                        if door_state[n] == 'closed':
                            open_door(door_state[n])
                            print('Door Opened:', n)
                            pygame.draw.circle(screen, "Yellow", (self.co(n)[0] * 80+500, self.co(n)[1] * 80), 20)
                            pygame.display.update()
                            clock.tick(2)
                            close_door(door_state[n])
                            print('Door Closed:', n)
                            pygame.draw.circle(screen, "Red", (self.co(n)[0] * 80+500, self.co(n)[1] * 80), 20)
                            pygame.display.update()
                            clock.tick(2)
                        elif door_state[n] == 'opened':
                            close_door(door_state[n])
                            print('Door Closed:', n)
                            pygame.draw.circle(screen, "Red", (self.co(n)[0] * 80+500, self.co(n)[1] * 80), 20)
                            pygame.display.update()
                            clock.tick(2)
                        
                for n in reconst_path:
                    pygame.draw.circle(screen, "White", (self.co(n)[0]*80+500, self.co(n)[1] * 80), 20)
                    pygame.display.update()
                    
                return end_node, orders, cups_available1, cups_carried1
            
                # for all neighbors of the current node do
            for (m, weight) in graph.get_neighbors(n):
                    # if the current node isn't in both open_list and closed_list
                    # add it to open_list and note n as it's parent
                    
                if m not in open_list and m not in closed_list:

                    if node_states[m] != 'reachable':
                        continue
                         
                    else:
                        open_list.add(m)
                        parents[m] = n
                        g[m] = g[n] + weight
                        
                        pygame.draw.circle(screen, "Blue", (self.co(m)[0]*80+500, self.co(m)[1] * 80), 20)
                        pygame.display.update()
                        clock.tick(2)

                    # otherwise, check if it's quicker to first visit n, then m
                    # and if it is, update parent data and g data
                    # and if the node was in the closed_list, move it to open_list
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n

                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)
                    
                # remove n from the open_list, and add it to closed_list
                # because all of his neighbors were inspected
            open_list.remove(n)
            closed_list.add(n)
            
        if len(open_list)==0:
            for loc in Map:
                if orders[0][0] == Map[loc]:
                    return '', orders, cups_available1, cups_carried1
        
        print('Path does not exist!!')
```
